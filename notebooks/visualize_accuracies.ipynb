```python
import sys
import os
import matplotlib.pylab as plt
import seaborn as sns
import glob

sys.path.append('..')

import ghiaseddin
import re
%matplotlib inline
```

```python
def plot_file(faddrs):
    sns.set_context("notebook")
    fig = plt.figure(figsize=(12, 6))
    ax = fig.add_subplot(111)
    ax.set_ylim(50, 100)
    ax.set_ylabel('accuracy (%)')
    ax.set_xlabel('epochs')
    
    if not isinstance(faddrs, list):
        faddrs = [faddrs]
    
    for faddr in faddrs:
        with open(faddr, 'r') as f:
            lines = f.readlines()
        lines = [float(l) for l in lines]
        ax.plot(lines)
    plt.show()
    plt.close()
```

```python
dataset = 'LFW10' #'Zap1', 'LFW10'
randomseed = '*' #'0', '1', '2', '3', '4'
```

```python
results = glob.glob(os.path.join(ghiaseddin.settings.result_models_root,
                                 'acc|baseline|*-d:%s-*-rs:%s-*' % (dataset, randomseed)))
print len(results)
```

```python
plot_file(results)
```
